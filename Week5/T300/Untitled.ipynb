```python
import numpy as np
import matplotlib.pyplot as plt

# Load replica temperature data
data = np.loadtxt("replica_temp.xvg")

# Extract time and temperature values
time = data[:, 0]
temperatures = data[:, 1:]

# Create a plot for each replica at every 100th frame
frame_interval = 100
for replica_id in range(temperatures.shape[1]):
    # Extract data at every 100th frame
    selected_temperatures = temperatures[::frame_interval, replica_id]
    selected_time = time[::frame_interval]

    plt.plot(selected_time, selected_temperatures, label=f"Replica {replica_id + 1}")

plt.xlabel("Time (ps)")
plt.ylabel("Temperature (K)")
plt.title("Temperature Trajectory of Each Replica (Every 100 Frames)")
plt.legend()
plt.grid(True)
plt.show()
```
