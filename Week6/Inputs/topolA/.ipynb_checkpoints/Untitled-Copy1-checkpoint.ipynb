```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md

# Load the trajectory data from the XTC file
xtc_file = "traj_comp.xtc"  # Replace with the actual XTC file name
topology_file = "dialaA.pdb"  # Replace with the actual topology file name
traj = md.load(xtc_file, top=topology_file)

# Compute the phi and psi dihedrals using MDTraj
phi_indices, phi_values = md.compute_phi(traj)
psi_indices, psi_values = md.compute_psi(traj)

# Plot phi vs psi
plt.figure(figsize=(8, 6))
plt.scatter(phi_values, psi_values, alpha=0.5)
plt.xlabel("Phi Dihedral (radians)")
plt.ylabel("Psi Dihedral (radians)")
plt.title("Ramachandran Plot")
plt.show()
```

```python
# Calculate the standard deviations
phi_std = np.std(phi_values, axis=0)
psi_std = np.std(psi_values, axis=0)

print("Standard Deviation of Phi Dihedral Angles:")
print(phi_std)
print("Standard Deviation of Psi Dihedral Angles:")
print(psi_std)
```
