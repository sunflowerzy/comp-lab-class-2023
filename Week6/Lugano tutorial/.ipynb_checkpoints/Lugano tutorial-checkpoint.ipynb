```python
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt

grid=np.loadtxt('lugano.grid.dat',comments=['#'])
colvar=np.loadtxt('COLVAR',comments=['#'])
hills=np.loadtxt('HILLS',comments=['#'])
```

```python
plt.scatter(colvar[:, 0][::10], colvar[:, 1][::10], s=1)
plt.xlabel('time(ns)')
plt.ylabel('Phi')
plt.show()
```

```python
plt.scatter(hills[:,0],hills[:,3],s=1)
plt.xlabel('time(ns)')
plt.ylabel('Gaussian hill height kj/mol')
plt.title('Time evolution of the Gaussian height')
plt.show()
```

```python
plt.plot(grid[:,0],-grid[:,2])
plt.title('Negative Energy from Grid File')
plt.xlabel('phi/radians')
plt.ylabel('free energy/ kj mol$^{-1}$')
plt.show()
```

```python
fes=np.loadtxt('fes.dat',comments=['#'])
plt.plot(fes[:,0],fes[:,1])
plt.xlabel('phi/radians')
plt.ylabel('free energy/ kj mol$^{-1}$')
plt.show()
```

```python
n01 = np.loadtxt('fes_1.dat', comments=['#'])
n03 = np.loadtxt('fes_3.dat', comments=['#'])
n10 = np.loadtxt('fes_10.dat', comments=['#'])
n20 = np.loadtxt('fes_20.dat', comments=['#'])
n50 = np.loadtxt('fes_50.dat', comments=['#'])
n100 = np.loadtxt('fes_100.dat', comments=['#'])

plt.plot(n01[:,0], n01[:,1], label='0.1ns')
plt.plot(n03[:,0], n03[:,1], label='0.3ns')
plt.plot(n10[:,0], n10[:,1], label='1ns')
plt.plot(n20[:,0], n20[:,1], label='2ns')
plt.plot(n50[:,0], n50[:,1], label='5ns')
plt.plot(n100[:,0], n100[:,1], label='10ns')
plt.xlabel('phi/radians')
plt.ylabel('free energy/ kj mol$^{-1}$')
plt.legend()
plt.show()
```
