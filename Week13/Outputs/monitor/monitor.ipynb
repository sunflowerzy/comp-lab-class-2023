```python
import numpy as np
import matplotlib.pyplot as plt

# Load energy files for QM/MM run
qm_energy_data = np.loadtxt('MONITOR-1.ener')
```

```python
# Extract time and potential energy for each run
time_qm, energy_qm = qm_energy_data[:, 1], qm_energy_data[:, 4]
```

```python
# Plot potential energy versus time 
plt.figure(figsize=(10, 6))
plt.plot(time_qm, energy_qm, label='QM/MM Run')
plt.xlabel('Time')
plt.ylabel('Potential Energy')
plt.title('Potential Energy vs Time for QM/MM')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the reaction coordinate data
monitor_data = np.loadtxt('MONITOR-COLVAR.metadynLog')

# Extract time and reaction coordinate
time_monitor, reaction_coord = monitor_data[:, 0], monitor_data[:, 1]

# Plot the reaction coordinate versus time
plt.figure(figsize=(10, 6))
plt.plot(time_monitor, reaction_coord)
plt.xlabel('Time')
plt.ylabel('Reaction Coordinate')
plt.title('Reaction Coordinate vs Time')
plt.grid(True)


# Show the plot
plt.show()
```
