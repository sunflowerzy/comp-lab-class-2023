```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load NVT energy files
nvt_energy_data = np.loadtxt('NVT-1.ener')
```

```python
# Extract time and potential energy
time_nvt, energy_nvt = nvt_energy_data[:, 1], nvt_energy_data[:, 4]
```

```python
# Plot potential energy vs time
plt.figure(figsize=(10, 6))

plt.plot(time_nvt, energy_nvt, label='NVT')
plt.xlabel('Time')
plt.ylabel('Potential Energy')
plt.title('Potential Energy vs Time')
plt.legend()
plt.grid(True)


# Show the plot
plt.show()
```

```python
npt_energy_data = np.loadtxt('NPT-1.ener')
time_npt, energy_npt = npt_energy_data[:, 0], npt_energy_data[:, 4]
# Plot potential energy vs time
plt.figure(figsize=(10, 6))
plt.plot(time_npt, energy_npt, label='NPT')
plt.xlabel('Time')
plt.ylabel('Potential Energy')
plt.title('Potential Energy vs Time')
plt.legend()
plt.grid(True)


# Show the plot
plt.show()
```
