```python
import numpy as np
import matplotlib.pyplot as plt

# Load energy files for QM/MM run
qm_energy_data = np.loadtxt('METADYN-1.ener')

# Extract time and potential energy for each run
time_qm, energy_qm = qm_energy_data[:, 1], qm_energy_data[:, 4]

# Plot potential energy versus time 
plt.figure(figsize=(10, 6))
plt.plot(time_qm, energy_qm, label='QM/MM Run')
plt.xlabel('Time')
plt.ylabel('Potential Energy')
plt.title('Potential Energy vs Time')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



# Load the reaction coordinate data from metad1_initmonitor
init_monitor_data = np.loadtxt('METADYN-COLVAR.metadynLog')

# Extract time and reaction coordinate from metad1_initmonitor
time_init_monitor, reaction_coord_init_monitor = init_monitor_data[:, 0], init_monitor_data[:, 1]

# Plot the reaction coordinate versus time
plt.figure(figsize=(10, 6))
plt.plot(time_init_monitor, reaction_coord_init_monitor, label='QM/MM Initial Monitor', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Reaction Coordinate')
plt.title('Reaction Coordinate vs Time')
plt.legend()
plt.grid(True)
```
