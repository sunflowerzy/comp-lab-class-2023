```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function to read XVG files
def read_xvg(file_path):
    data = {"x": [], "y": []}
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            parts = line.strip().split()
            if len(parts) == 2:
                x, y = map(float, parts)
                data["x"].append(x)
                data["y"].append(y)
    return data

# Now you can use matplotlib to create plots
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function to read XVG files
def read_xvg(file_path):
    data = {"x": [], "y": []}
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            parts = line.strip().split()
            if len(parts) == 2:
                x, y = map(float, parts)
                data["x"].append(x)
                data["y"].append(y)
    return data

# Specify the path to your XVG file
xvg_file_path = "/home/zg2344/comp-lab-class/comp-lab-class-2023/Week3/potential.xvg"

# Read data from the XVG file
data = read_xvg(xvg_file_path)

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(data["x"], data["y"], label="Data")
plt.xlabel("Energy Minimization")
plt.ylabel("Potential Energy (KJ mol^-1)")
plt.title("Energy Minimization")
plt.grid(True)

# Show the plot in the Jupyter Notebook
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the energy file
data = np.loadtxt("temperature.xvg", comments=('@', '#'))

# Extract the time and temperature values
time = data[:, 0]  # Time in ps
temperature = data[:, 1]  # Temperature in K

# Create a plot for temperature vs. time
plt.figure(figsize=(10, 6))
plt.plot(time, temperature, label="Temperature")
plt.xlabel("Time (ps)")
plt.ylabel("Temperature (K)")
plt.title("Temperature vs. Time")
plt.grid(True)
plt.legend()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Read the data from the first RMSD file (rmsd_xtal.xvg)
data1 = np.loadtxt("rmsd_xtal.xvg", comments=('@', '#'))
time_ns1 = data1[:, 0]  # Time in ns
rmsd_nm1 = data1[:, 1]  # RMSD in nm

# Read the data from the second RMSD file (rmsd.xvg)
data2 = np.loadtxt("rmsd.xvg", comments=('@', '#'))
time_ns2 = data2[:, 0]  # Time in ns
rmsd_nm2 = data2[:, 1]  # RMSD in nm

# Create a plot for RMSD vs. time for the first data
plt.figure(figsize=(12, 6))
plt.plot(time_ns1, rmsd_nm1, label="RMSD (Crystal)")
plt.plot(time_ns2, rmsd_nm2, label="RMSD (equilibrated)")
plt.xlabel("Time (ns)")
plt.ylabel("RMSD (nm)")
plt.title("RMSD vs. Time")
plt.grid(True)
plt.legend()
plt.show()
```
