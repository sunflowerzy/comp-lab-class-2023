```python
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis.rdf import InterRDF
import MDAnalysis as mda
```

```python
Step, Time, Kin, Temp, Pot, ConsQty, UsedTime = np.loadtxt('proton_aimd_1ps_every0.5fs-1.ener', skiprows=1, unpack=True)
```

```python
plt.plot(Time, Pot)
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u.)')
plt.title('Potential Energy vs. Time for proton')

plt.show()
```

```python
plt.plot(Time, Kin)
plt.xlabel('Time (fs)')
plt.ylabel('kinetic Energy (a.u.)')
plt.title('kinetic Energy vs. Time for proton')

plt.show()
```

```python
traj = mda.Universe('proton_aimd_1ps_every0.5fs-pos-1.xyz')
traj

traj.dimensions = np.array([12.42, 12.42, 12.42, 90, 90, 90])
```

```python
box_size = 12.42
rdf = InterRDF(traj.select_atoms('name O'), traj.select_atoms('name O'), range=(0.0, box_size / 2), nbins=100)
rdf.run()
```

```python
plt.plot(rdf.bins[1:], rdf.rdf[1:])

plt.xlabel('Distance (Angstrom)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function for Oxygen in system of solvated proton in water')



plt.show()
```
